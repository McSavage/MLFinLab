```python
import pandas as pd
data = pd.read_csv("https://raw.githubusercontent.com/hudson-and-thames/example-data/main/tick_data.csv")
data
```

```python
# Don't convert to datetime here -- computation will take long due to the size of the tick data files
date_time = data["Date"] + " " + data["Time"]
new_data = pd.concat([date_time, data["Price"], data["Volume"]], axis=1)
new_data.columns = ["date", "price", "volume"]
new_data
```

```python
# Save to csv in the data folder
import os

# Create path to data folder (one level up from notebooks)
data_folder = os.path.join("..", "data")
output_path = os.path.join(data_folder, "processed_tick_data.csv")

# Create data folder if it doesn't exist
os.makedirs(data_folder, exist_ok=True)

# Save the processed data
new_data.to_csv(output_path, index=False)
print(f"Data saved to: {output_path}")
print(f"File size: {len(new_data)} rows")
```
